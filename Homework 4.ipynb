Now, features that I am going to add for each team would be:

**For last 6 matches**


1. mean_home_goals

7. Team Point

I would like my algorithm to know which team is stronger and which weak. So I would give every team a score in every league according to their win counts on our total data and scale the scores in each league between **1** and **10**.

```python
def calculate_team_points(match_id, league_id):
    
    
    matches2 = matches[matches['league_id'] == league_id].reset_index(drop=True)
    teams_list = matches2.home_id.unique().tolist()
    
    team_points = dict()
    for i in teams_list:
        team_points[i] = 0
    
    
    winners2 = matches2.winners.tolist()
    
    for i in teams_list:
        team_points[i] = winners2.count(str(i))
    
    table = pd.DataFrame(zip(team_points.keys(), team_points.values()), columns = ["teams", 'points'])
    table = table.sort_values(by = 'points', ascending= False).reset_index(drop=True)
    
    x = table[['points']].values.astype(float)
    
    
    # Scaling points between 1 - 10
    news = list()
    for k in x:
        rng = max(x)[0] - min(x)[0] 
        new = 1 + abs(k -  min(x)[0])*(9/rng)
        news.append(new[0])
        
   
    team_points_test = table.copy()
    team_points_test['points'] = news
    return team_points_test
```

```python
def get_team_rates(match_id, league_id, team_id, order=1):
    
    league_games = matches[matches['league_id'] == league_id]
    index = league_games[league_games['match_id'] == match_id ].index[0]
    league_games_played = league_games.loc[:index,].copy()
    
    games_played = league_games_played[(league_games_played.away_id == team_id) | (league_games_played.home_id == team_id)].reset_index(drop=True)
    
    if len(games_played) < 7: # We need at least 6 prior games to evaluate performance in the past.
        return None
    
    
    
    

    
    away_games = games_played[(games_played.away_id == team_id)].reset_index(drop=True)
   
    #print("Away_games : " , away_games.match_id.tolist())
    away_goals = away_games.tail(6).away_score.tolist()
    
    mean_away_goals = statistics.mean(away_goals)           #*

      
    
    home_games = games_played[(games_played.home_id == team_id)].reset_index(drop=True)
   
    
    home_goals = home_games.tail(6).away_score.tolist()
   
    mean_home_goals = statistics.mean(home_goals)           #*

    
    away_conceded = away_games.tail(6).home_score.tolist()
    home_conceded = home_games.tail(6).away_score.tolist()

    mean_away_conceded = statistics.mean(away_conceded)     #*


    mean_home_conceded = statistics.mean(home_conceded)     #*
 
    mean_goals = sum(home_goals + away_goals) / len(home_goals + away_goals)    #*
       
    
    stdev_goals = statistics.stdev(list(home_goals) + list(away_goals))     #*
  
    #print("oldest match : " , games_played_2.match_id.values)
    oldest_match = games_played.match_id.values[0]
    team_pts = calculate_team_points(oldest_match, league_id)
    team_pt = team_pts[team_pts['teams'] == team_id]['points'].values[0]           #*
    
    
    
    if order == 1:
        value = {'mean_away_goals' : mean_away_goals, 'mean_home_goals': mean_home_goals, 'mean_away_conceded': mean_away_conceded,
                 'mean_home_conceded': mean_home_conceded, 'mean_goals': mean_goals, 'stdev_goals':stdev_goals, 'team_pt': team_pt}

    else: 
        value = {'mean_away_goals2' : mean_away_goals, 'mean_home_goals2': mean_home_goals, 'mean_away_conceded2': mean_away_conceded,
                 'mean_home_conceded2': mean_home_conceded, 'mean_goals2': mean_goals, 'stdev_goals2':stdev_goals, 'team_pt2':team_pt}
    
    
    return value
    
```

So lets start creating our dataset which we will run our model on:

```python
rows = list()
with tqdm(total = len(matches.match_id.tolist())) as pbar:
    for i in matches.match_id.index:

        match_id = matches.at[i,'match_id']
        home_id = matches.at[i, 'home_id']
        away_id = matches.at[i, 'away_id']
        league_id = matches.at[i, 'league_id']
        total_goals = matches.at[i, 'total_goals']
        bigger = matches.at[i, 'bigger']
        
        main = {'match_id' : match_id , "home_id": home_id, "away_id": away_id, 'total_goals':total_goals, 'bigger': bigger}

        dt = get_team_rates(match_id = match_id, league_id = league_id,   team_id = home_id)
        if dt == None:
            pbar.update(1)
            continue

        dt2 = get_team_rates(match_id = match_id, league_id = league_id, team_id = away_id, order=2)
        if dt2 == None:
            pbar.update(1)
            continue
        main.update(dt)
        main.update(dt2)
        

        rows.append(main)
        pbar.update(1)
    
df = pd.DataFrame(rows)
df.head()
```

## 2) Let's try to find the best parameters for each Machine Learning method:

###  I) Decision Tree with min_samples_leaf = 0.1 with 10 fold cross validation
